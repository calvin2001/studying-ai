```python
import numpy as np
import matplotlib.pyplot as plt

# defining vectors
a = np.array([2, 1])
b = np.array([-1, 2])

# calculating dot product
dot = np.dot(a, b)
print("Vector a · b =", dot)

# visualization function
def draw_2vectors(v1, v2):
    plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')
    plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='b')
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.legend()

#  visualization
plt.figure()
draw_2vectors(a, b)
plt.title("Vector Dot Product Visualization")
plt.show()
```
