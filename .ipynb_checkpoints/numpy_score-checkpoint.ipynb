```python
import numpy as np

# í•™ìƒ ì„±ì  ë°ì´í„° (5ëª…ì˜ í•™ìƒ, 4ê³¼ëª© ì ìˆ˜)
# í–‰: í•™ìƒ, ì—´: [ìˆ˜í•™, ì˜ì–´, ê³¼í•™, ì‚¬íšŒ]
scores = np.array([
    [85, 92, 78, 88],  # í•™ìƒ 1
    [90, 87, 95, 92],  # í•™ìƒ 2
    [78, 85, 82, 79],  # í•™ìƒ 3
    [92, 88, 90, 94],  # í•™ìƒ 4
    [87, 90, 85, 86]   # í•™ìƒ 5
])

subjects = ['ìˆ˜í•™', 'ì˜ì–´', 'ê³¼í•™', 'ì‚¬íšŒ']
students = ['í•™ìƒ1', 'í•™ìƒ2', 'í•™ìƒ3', 'í•™ìƒ4', 'í•™ìƒ5']

print("ğŸ“ í•™ìƒ ì„±ì  ë¶„ì„ ì‹œìŠ¤í…œ")
print("=" * 50)

# 1. ê¸°ë³¸ ì •ë³´
print(f"í•™ìƒ ìˆ˜: {scores.shape[0]}ëª…")
print(f"ê³¼ëª© ìˆ˜: {scores.shape[1]}ê³¼ëª©")
print()

# 2. ê° í•™ìƒì˜ í‰ê·  ì ìˆ˜
student_averages = np.mean(scores, axis=1)
print("ğŸ“Š í•™ìƒë³„ í‰ê·  ì ìˆ˜:")
for i, avg in enumerate(student_averages):
    print(f"{students[i]}: {avg:.2f}ì ")
print()

# 3. ê° ê³¼ëª©ì˜ í‰ê·  ì ìˆ˜
subject_averages = np.mean(scores, axis=0)
print("ğŸ“ˆ ê³¼ëª©ë³„ í‰ê·  ì ìˆ˜:")
for i, avg in enumerate(subject_averages):
    print(f"{subjects[i]}: {avg:.2f}ì ")
print()

# 4. ìµœê³  ì ìˆ˜ì™€ ìµœì € ì ìˆ˜
max_score = np.max(scores)
min_score = np.min(scores)
max_position = np.unravel_index(np.argmax(scores), scores.shape) // 7.20: ì¬êµ¬í˜„
min_position = np.unravel_index(np.argmin(scores), scores.shape)

print(f"ğŸ† ìµœê³  ì ìˆ˜: {max_score}ì ")
print(f"   ìœ„ì¹˜: {students[max_position[0]]}, {subjects[max_position[1]]}")
print()
print(f"ğŸ“‰ ìµœì € ì ìˆ˜: {min_score}ì ")
print(f"   ìœ„ì¹˜: {students[min_position[0]]}, {subjects[min_position[1]]}")
print()

# 5. 90ì  ì´ìƒ ë°›ì€ ê³¼ëª© ìˆ˜ (í•™ìƒë³„)
high_scores = scores >= 90
high_score_counts = np.sum(high_scores, axis=1)
print("ğŸŒŸ 90ì  ì´ìƒ ê³¼ëª© ìˆ˜ (í•™ìƒë³„):")
for i, count in enumerate(high_score_counts):
    print(f"{students[i]}: {count}ê³¼ëª©")
print()

# 6. ê° ê³¼ëª©ë³„ í‘œì¤€í¸ì°¨ (ë‚œì´ë„ ì¸¡ì •)
subject_stds = np.std(scores, axis=0)
print("ğŸ“Š ê³¼ëª©ë³„ í‘œì¤€í¸ì°¨ (ë‚®ì„ìˆ˜ë¡ ì ìˆ˜ í¸ì°¨ ì‘ìŒ):")
for i, std in enumerate(subject_stds):
    print(f"{subjects[i]}: {std:.2f}")
print()

# 7. í•™ê¸‰ í†µê³„
class_average = np.mean(scores)
class_std = np.std(scores)
print("ğŸ« í•™ê¸‰ ì „ì²´ í†µê³„:")
print(f"ì „ì²´ í‰ê· : {class_average:.2f}ì ")
print(f"ì „ì²´ í‘œì¤€í¸ì°¨: {class_std:.2f}")
print()

# 8. ìš°ìˆ˜ í•™ìƒ ì°¾ê¸° (í‰ê·  90ì  ì´ìƒ)
excellent_students = student_averages >= 90
print("â­ ìš°ìˆ˜ í•™ìƒ (í‰ê·  90ì  ì´ìƒ):")
for i, is_excellent in enumerate(excellent_students):
    if is_excellent:
        print(f"{students[i]}: {student_averages[i]:.2f}ì ")

# 9. ê³¼ëª©ë³„ ìˆœìœ„ (ìˆ˜í•™ ê³¼ëª© ì˜ˆì‹œ)
math_scores = scores[:, 0]
math_ranking = np.argsort(math_scores)[::-1]  # ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
print(f"\nğŸ¥‡ ìˆ˜í•™ ê³¼ëª© ìˆœìœ„:")
for rank, student_idx in enumerate(math_ranking):
    print(f"{rank+1}ë“±: {students[student_idx]} ({math_scores[student_idx]}ì )")

# 10. ë³´ë„ˆìŠ¤: ì„±ì  í–¥ìƒ ì‹œë®¬ë ˆì´ì…˜
print(f"\nğŸ¯ ëª¨ë“  í•™ìƒì´ 5ì ì”© í–¥ìƒí–ˆë‹¤ë©´:")
improved_scores = scores + 5
improved_average = np.mean(improved_scores)
print(f"í–¥ìƒ í›„ í•™ê¸‰ í‰ê· : {improved_average:.2f}ì  (í˜„ì¬: {class_average:.2f}ì )")
print(f"í‰ê·  ìƒìŠ¹í­: {improved_average - class_average:.2f}ì ")
```
