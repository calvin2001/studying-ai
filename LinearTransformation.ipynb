```python
import numpy as np
import matplotlib.pyplot as plt

# basis vectors
i = np.array([1, 0])
j = np.array([0, 1])

# linear transformation matrix A
A = np.array([[2, 1],
              [1, 3]])

# transformed vector
i_trans = A @ i
j_trans = A @ j

# vector visualizing function
def draw_vectors(vecs, colors, labels):
    for vec, color, label in zip(vecs, colors, labels):
        plt.quiver(0, 0, vec[0], vec[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.grid()
    plt.gca().set_aspect('equal')
    plt.legend()

# visualization
plt.figure()
draw_vectors([i, j], ['r', 'b'], ['i', 'j'])             # the original basis
draw_vectors([i_trans, j_trans], ['orange', 'green'], ['A*i', 'A*j'])  # transformed basis
plt.title("linear transformation: change of the standard basis vectors")
plt.show()
```
