```python
import numpy as np
import matplotlib.pyplot as plt

# 간단한 2차 함수 f(x) = x^2
def f(x):
    return x**2

# 도함수 f'(x) = 2x
def df_dx(x):
    return 2*x

# 그래프로 시각화
x = np.linspace(-5, 5, 100)
y = f(x)
dy = df_dx(x)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b-', label='f(x) = x²')
plt.title(' original function')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, dy, 'r-', label="f'(x) = 2x")
plt.title('derivitive')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 2변수 함수 f(x,y) = x² + y²
def f_2d(x, y):
    return x**2 + y**2

# 편미분
def df_dx_2d(x, y):
    return 2*x

def df_dy_2d(x, y):
    return 2*y

# 그래디언트 벡터
def gradient(x, y):
    return np.array([df_dx_2d(x, y), df_dy_2d(x, y)])

# 특정 점에서의 그래디언트
point = (3, 4)
grad = gradient(3, 4)
print(f"gradient of the point{point}: {grad}")
```
