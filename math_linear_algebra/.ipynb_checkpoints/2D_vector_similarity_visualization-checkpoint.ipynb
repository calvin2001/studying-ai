```python
import numpy as np
import matplotlib.pyplot as plt

def vector_angle(v1, v2):
    cos_theta = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    angle = np.arccos(np.clip(cos_theta, -1.0, 1.0))
    return np.degrees(angle)

# user defining vectors
a = np.array([1, 2])
b = np.array([2, 1])

# calculating dot product, angle
dot = np.dot(a, b)
angle = vector_angle(a, b)

# visualization
plt.figure()
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='b')
plt.title(f"angle: {angle:.2f}°, dot product: {dot}")
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.grid()
plt.legend()
plt.gca().set_aspect('equal')
plt.show()
```
