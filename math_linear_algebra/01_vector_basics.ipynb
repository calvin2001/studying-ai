```python
import numpy as np
import matplotlib.pyplot as plt

def draw_vector(v, color='r', label=None):
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=color, label=label)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.grid()
    if label:
        plt.legend()

v1 = np.array([3, 4])
v2 = np.array([1, -2])

plt.figure()
draw_vector(v1, 'r', 'v1')
draw_vector(v2, 'b', 'v2')
draw_vector(v1 + v2, 'g', 'v1+v2')
plt.title("Vector Addition")
plt.show()
```

```python
s = 2
v_scaled = s * v1

plt.figure()
draw_vector(v1, 'r', 'v1')
draw_vector(v_scaled, 'g', '2*v1')
plt.title("Scalar Multiplication")
plt.show()
```
